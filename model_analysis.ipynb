##### Accuracy table

```python
accuracy_table = {
    "Model Name": list(),
    "Epochs": list(),
    "Avg Train Acc": list(),
    "Std Train Acc": list(),
    "Best Train Acc": list(),
    "Avg Val Acc": list(),
    "Std Val Acc": list(),
    "Best Val Acc": list()
}

def add_acc_baseline(baseline_results):
    for baseline in baseline_results:
        baseline_name = baseline.split("\\")[-1].replace("_result.csv", "")
        baseline_df = pd.read_csv(baseline)

        num_epochs = baseline_df["Epoch"].values.max()

        train_acc = baseline_df["Training Accuracy"].values
        val_acc = baseline_df["Validation Accuracy"].values

        avg_train_acc = round(train_acc.mean(), 3)
        std_train_acc = round(train_acc.std(), 3)
        best_train_acc = round(train_acc.max(), 3)

        avg_val_acc = round(val_acc.mean(), 3)
        std_val_acc = round(val_acc.std(), 3)
        best_val_acc = round(val_acc.max(), 3)

        accuracy_table["Model Name"].append(baseline_name)
        accuracy_table["Epochs"].append(num_epochs)
        accuracy_table["Avg Train Acc"].append(avg_train_acc)
        accuracy_table["Std Train Acc"].append(std_train_acc)
        accuracy_table["Best Train Acc"].append(best_train_acc)

        accuracy_table["Avg Val Acc"].append(avg_val_acc)
        accuracy_table["Std Val Acc"].append(std_val_acc)
        accuracy_table["Best Val Acc"].append(best_val_acc)

def add_acc_pbn(pbn_results):
    for pbn_result in pbn_results:
        pbn_name = pbn_result.split("\\")[-1].replace("_result.csv", "")
        pbn_df = pd.read_csv(pbn_result)

        num_epochs = pbn_df["Epoch"].values.max()

        train_acc = pbn_df[pbn_df['Mode'] == 'train']["Accuracy"].values
        val_acc = pbn_df[pbn_df['Mode'] == 'validation']["Accuracy"].values

        avg_train_acc = round(train_acc.mean(), 3)
        std_train_acc = round(train_acc.std(), 3)
        best_train_acc = round(train_acc.max(), 3)

        avg_val_acc = round(val_acc.mean(), 3)
        std_val_acc = round(val_acc.std(), 3)
        best_val_acc = round(val_acc.max(), 3)

        accuracy_table["Model Name"].append(pbn_name)
        accuracy_table["Epochs"].append(num_epochs)
        accuracy_table["Avg Train Acc"].append(avg_train_acc)
        accuracy_table["Std Train Acc"].append(std_train_acc)
        accuracy_table["Best Train Acc"].append(best_train_acc)

        accuracy_table["Avg Val Acc"].append(avg_val_acc)
        accuracy_table["Std Val Acc"].append(std_val_acc)
        accuracy_table["Best Val Acc"].append(best_val_acc)

add_acc_baseline(baseline_results)
add_acc_pbn(pbn_vgg16_results)
add_acc_pbn(pbn_resnet18_results)
add_acc_pbn(mpbn_results)
accuracy_table = pd.DataFrame(accuracy_table)
    
```

##### Case 1: mPBN failed to learn giraffe and zebra, why?

Issues: <br>
For mPBN with 30 classes:
- When using 10 and 30 prototypes, it cannot classify giraffe
- When using 20 prototypes, it cannot classify zebra
- Same issue did NOT occur when using Resnet as PBN backbone

```python
with open("dataset_infos/singleLabel_train_data_30classes.json", "r") as f:
    train_data_30cls = json.load(f)
with open("dataset_infos/singleLabel_val_data_30classes.json", "r") as f:
    val_data_30cls = json.load(f)

giraffe_ids_train = set([k for k,v in train_data_30cls['img_labels'].items() if v == 'giraffe'])
giraffe_ids_val = set([k for k,v in val_data_30cls['img_labels'].items() if v == 'giraffe'])

zebra_ids_train = set([k for k,v in train_data_30cls['img_labels'].items() if v == 'zebra'])
zebra_ids_val = set([k for k,v in val_data_30cls['img_labels'].items() if v == 'zebra'])

print(f"How many giraffes are there in train? {len(giraffe_ids_train)}")
print(f"How many giraffes are there in val? {len(giraffe_ids_val)}")

print(f"How many zebras are there in train? {len(zebra_ids_train)}")
print(f"How many zebras are there in val? {len(zebra_ids_val)}")
    
```

```python
### Are they leaked within class?
print(f"Is there any overlap between train and val for giraffes? \
      {len(giraffe_ids_train.intersection(giraffe_ids_val)) > 0}")
print(f"Is there any overlap between train and val for zebras? \ 
      {len(zebra_ids_train.intersection(zebra_ids_val)) > 0}")

# The result said no
```

```python
### Are they leaked between class?
print(f"Is there any overlap between train giraffes and zebras? \
      {len(giraffe_ids_train.intersection(zebra_ids_train)) > 0}")
print(f"Is there any overlap between val giraffes and zebras? \
      {len(giraffe_ids_val.intersection(zebra_ids_val)) > 0}")

# The result said no
```

```python
mpbn_10ptypes_cls_labels = dict()

for i, label in enumerate(mpbn_10ptypes_confmatrix.columns):
    mpbn_10ptypes_cls_labels[label] = i - 1

# check which index is giraffe and which one is zebra
giraffe_idx = str(mpbn_10ptypes_cls_labels['giraffe'])
zebra_idx = str(mpbn_10ptypes_cls_labels['zebra'])
print(f"Giraffe index: {giraffe_idx}, zebra index: {zebra_idx}")

giraffe_giraffe = f'{giraffe_idx}_{giraffe_idx}'
giraffe_zebra = f'{giraffe_idx}_{zebra_idx}'
zebra_giraffe = f'{zebra_idx}_{giraffe_idx}'
zebra_zebra = f'{zebra_idx}_{zebra_idx}'
```

```python
# check which ones are which
for pred_true_pair in [giraffe_giraffe, giraffe_zebra, zebra_giraffe, zebra_zebra]:
    if pred_true_pair not in mpbn_10ptypes_result['confusion_mapping'].keys():
        print(f"Prediction for {pred_true_pair} is zero.")
    else:
        print(f"Prediction for {pred_true_pair} is {mpbn_10ptypes_result['confusion_mapping'][pred_true_pair]}")
```
